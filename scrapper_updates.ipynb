```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import re


regex = re.compile('[1234567890]')
dd = {"title":[], "price":[], "rating":[], "reviews":[],"ASIN":[],"Manufacturer":[],"URL":[],"description":[],}

def abc(page):
    
   
    HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36', 'Accept-Language':'en-US, en;q=0.5'})

    URL = page
    
    webpage = requests.get(URL, headers=HEADERS)

   
    soup = BeautifulSoup(webpage.content, "html.parser")


    links = soup.find_all("a", attrs={'class':'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'})

   
    links_list = []

   
    for link in links:
            links_list.append(link.get('href'))

    
    
     
    for link in links_list:
        new_webpage = requests.get("https://amazon.in" + link, headers=HEADERS)
        link_name = "https://amazon.in" + link
        new_soup = BeautifulSoup(new_webpage.content, "html.parser")

        dd['title'].append(get_title(new_soup))
        dd['price'].append(get_price(new_soup))
        dd['rating'].append(get_rating(new_soup))
        dd['reviews'].append(get_review_count(new_soup))
        dd['ASIN'].append(get_asin(new_soup))
        dd['Manufacturer'].append(get_manufacturer(new_soup))
        dd['URL'].append(link_name)
        dd['description'].append(get_description(new_soup))



        
        
        
        
        
        
def get_description(soup):

    try:
    
        des_string =soup.find("ul", attrs={"class":"a-unordered-list a-vertical a-spacing-mini"}).text.strip()

    except AttributeError:
        des_string = ""

    return des_string
        
        
        
        
def get_title(soup):

    try:
    
        title_string = soup.find("span", attrs={"id":'productTitle'}).text.strip()

    except AttributeError:
        title_string = ""

    return title_string


def get_price(soup):

    try:
        price = soup.find("span", attrs={"class":"a-offscreen"}).text


    except AttributeError:  
        price = ""

    return price


def get_rating(soup):

    try:
        rating = soup.find("i", attrs={"class":'a-icon a-icon-star a-star-4'}).text
    
    except AttributeError:
    
        rating = ""

    return rating


def get_review_count(soup):
    try:
        review_count = soup.find("span", attrs={"id":'acrCustomerReviewText'}).text

    except AttributeError:
        review_count = ""

    return review_count


def get_asin(soup):
    try:
        asin =soup.find("div", attrs={"id":'title_feature_div'})["data-csa-c-asin"]

    except AttributeError:
        asin = "Not Available"

    return asin        
        


def get_manufacturer(soup):
    try:
        manufacturer = soup.find("ul", attrs={"class":'a-unordered-list a-nostyle a-vertical a-spacing-none detail-bullet-list'}).find_all("li")[3]
        manufacturer = manufacturer.find_all("span")[2].text
        
        if(regex.search(manufacturer) == None):
            manufacturer=manufacturer
        else:
            manufacturer = soup.find("ul", attrs={"class":'a-unordered-list a-nostyle a-vertical a-spacing-none detail-bullet-list'}).find_all("li")[2]
            manufacturer = manufacturer.find_all("span")[2].text
        
    except AttributeError:
        manufacturer = ""

    return manufacturer





```
